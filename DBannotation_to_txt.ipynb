```python
import pandas as pd
import numpy as np
import glob,os, re
```

```python
with open('./datasets/AW_20220328/result.txt','r') as f:
    lines=f.readlines()
lines
```

```python
train = os.listdir('./datasets/AW_20220328/train2017/')
val = os.listdir('./datasets/AW_20220328/val2017/')

print(len(train), len(val))
```

```python
train_line=[]
val_line=[]
for line in lines:
    fname = line.split(',')[0]
    if fname in train:
        train_line.append(line)
    elif fname in val:
        val_line.append(line)

print(len(train_line), len(val_line))
```

```python
txt=[]
for i in range(len(train_line)):
    fname = train_line[i].split(',')[0]
    tmp = train_line[i].split('\"shapeCoord\":')
    for j in range(1,len(tmp)):
        tmp2 = tmp[j].split(',')
        xyxy=[]
        for k in range(4):
            #print('xyxy = ',int(re.sub(r'[^0-9]', '', tmp2[k])))
            xyxy.append(re.sub(r'[^0-9]', '', tmp2[k]))
    
        txt.append(fname+','+xyxy[0]+','+xyxy[1]+','+xyxy[2]+','+xyxy[3]+'\n')
print(len(txt))
```

```python
txt=[]
for i in range(len(val_line)):
    fname = val_line[i].split(',')[0]
    tmp = val_line[i].split('\"shapeCoord\":')
    for j in range(1,len(tmp)):
        tmp2 = tmp[j].split(',')
        xyxy=[]
        for k in range(4):
            #print('xyxy = ',int(re.sub(r'[^0-9]', '', tmp2[k])))
            xyxy.append(re.sub(r'[^0-9]', '', tmp2[k]))
    
        txt.append(fname+','+xyxy[0]+','+xyxy[1]+','+xyxy[2]+','+xyxy[3]+'\n')
print(len(txt))
```

```python
with open('./datasets/AW_20220328/val.txt','w') as f:
    for i in range(len(txt)):
        f.write(txt[i])
```
